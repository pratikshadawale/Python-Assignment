```python
import pandas as pd
```

```python
flights=pd.read_csv('flights.csv')
flights.head()
```

1. How many total number of days does the flights table cover?

```python
total_days = flights['day'].sum()
total_days
```

Ans : There are total 5291016 number of days in flight table.

```python
airports = pd.read_csv('airports.csv')
airports.head()
```

```python
airports.rename(columns={'IATA_CODE':'origin'}, inplace=True)
airports
```

```python
flights['CITY'] = flights.origin.map(airports.set_index('origin')['CITY'])
flights.head()
```

2. How many departure cities (not airports) does the flights database cover?

```python
dep_cities = flights['CITY'].count()
dep_cities
```

Ans : There are 336776 departure cities in flight table.

```python
planes=pd.read_csv('planes.csv')
planes.head()
```

3. What is the relationship between flights and planes tables?

Ans : Many to One relationship between flights and planes tables.

4. Which airplane manufacturer incurred the most delays in the analysis period?

```python
flights['manufacturer']= flights.tailnum.map(planes.set_index('tailnum')['manufacturer'])
flights.head()
```

```python
flights.groupby('manufacturer')['arr_delay'].count().reset_index(name='most_no_of_delays')\
                                 .sort_values(['most_no_of_delays'], ascending=False).head(1)
```

Ans : Manufacturer Boeing incurred the most delays in the analysis period. 

5. Which are the two most connected cities?

```python
flights.groupby('CITY')['flight'].count().reset_index(name='MOST_NO_OF_FLIGHTS')\
                                 .sort_values(['MOST_NO_OF_FLIGHTS'], ascending=False)
```

Ans : Since New York and Newark has the most number of flights that means they are the 2 most connected         cities.
